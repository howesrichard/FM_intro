# FINMA FM Equity Research Exercise

You are a 3-person equity research analyst team and you have been tasked with developing and implementing a program that produces an equity research report as a pdf file for a company selected by the user of your program.

One team member will be the leader, who will:
- create a new **private** GitHub repo called "finma_EquityAnalyst_reportGen" 
- invite 2 team members as collaborators
- copy the EAR_generator.ipynb from howesrichard/finma_FM_working into the new repo
- create a folder called "assets"
- push these changes to the main branch of the repo

The second team member will be responsible for the commentary that goes into the report:
- create a new branch on the leader/finma_EquityAnalyst_reportGen repo called "Commentary_function"
- create 3 text files named BHP.txt, CBA.txt, and GYG.txt, in the assets folder, which each contains a paragraph of text which will form the commentary that will get included into the final research report;
- create a new python file called "commentary.py" containing a function which returns a paragraph of commentary depending on the argument of the function which will the the three letter stock symbol 
- the function will open the appropriate text file read the text and return this text as the return value of the function
- once you are happy with you commentary.py file and your .txt files:
  - push your committed changes to the Commentary_funciton branch
  - create a pull request to merge the Commentary_function branch onto the main branch 

The third team member will be responsible for creating three historical stock price charts using the matplotlib library.
- create a new branch on the leader/finma_EquityAnalyst_reportGen repo called "Charts"
- create a jupyter notebook that:
  - imports the relevant libraries
  - gets stock price data from a csv file that you create or, if you are feeling adventurous, and can figure out how to do it, from data you call from the yfinance API
  - produces a historical stock price chart
- manually save this chart as a picture (.png for example)
- once you are happy with your charts saved as .png files into the assets folder:
  - push your committed changes to the Charts branch
  - create a pull request to merge the Charts branch onto the main branch 

The leader will then:
- review the pull requests
- merge the changes
- write code in the EAR_generator.ipynb to import and call the commentary function
- incorporate the commentary in the pdf output using fpdf
- incorporate the correct chart in the pdf output using fpdf
- produce the final pdf using fpdf
