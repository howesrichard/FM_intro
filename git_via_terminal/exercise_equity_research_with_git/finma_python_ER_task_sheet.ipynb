# FINMA FM Equity Research Exercise
### Task:
You are an equity research analyst and you have been tasked with developing and implementing a program that produces an equity research report as a pdf file for an assigned company in the ASX200. As a group, you will be randomly assigned a ticker to report on.

One team member will be the leader, who will create a new **private** GitHub repo and invite team members as collaborators, with each team member creating a different enhancement to the program which will become included in the research report pdf. In other words, the team should allocate each section of the report to a single member. Do not set your repo to public visibility at any time, before or after submission.

In your blank repo create a new notebook (.ipynb) file named _analyst\_report.ipynb_. You will develop your main program inside this notebook file.

### Features
The program should produce an attractive and well-formatted report. The features that your program can have, along with the marks assigned for each feature are:
- Create a pdf file of the final research report
- Key ratios and financial metrics informed through the use of an API, and computed in functions that have been created, saved into, and imported from a .py file
- Include commentary imported from a .txt file
- An extract from a DCF model built in Excel
- A chart showing price history for the stock, historical price targets of the research (taken from a CSV file), and the history of the ASX200
- A relevant image or table scraped from a website for the company

The focus of the task is on your ability to use Python and GitHub collaboratively, not on your financial analysis.


* Make sure all of your files are in a single folder. This folder should include:
  - Your main notebook file named analyst_report.ipynb.
  - Any .py files you used to store functions for importing to your main program.
  - Any .txt or .csv files that your main program needs to draw from.
  - The .xlsx file that contains your DCF analysis.
  - Any output files generated by running your program, including the final pdf equity report.
